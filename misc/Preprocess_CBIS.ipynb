```python
from PIL import Image
import numpy as np
Image_List=[]
Labels_List=[]
for image in imagePaths:
    # Load the Image , Resize it 224,224 pixels
    Im=np.asarray(Image.open(image).resize((200,200)))
    Im=Im.astype(dtype='float16')
    Image_List.append(Im)
    
    # Create a list of labels ['benign','malignant'] 
    Labels_List.append((image.partition(os.path.sep)[-1]).split('/')[-2])
    
data=np.array(Image_List)
label=np.array(Labels_List)
print(data.shape)
print(label.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
labels=lb.fit_transform(label)
print(type(labels))
print(labels[0])
print(label[0])
print(labels.shape)


import keras
labels=keras.utils.to_categorical(labels, num_classes=2, dtype='float32')
```

```python
print(labels.shape)
```

```python
from sklearn.model_selection import train_test_split
(trainX,testX,trainY,testY)=train_test_split(data,labels,test_size=0.2)


print('[Training data size]',len(trainX))
print('[Validation data size]',len(testX))
```

```python
from keras.preprocessing.image import ImageDataGenerator
trainX=trainX.reshape(len(trainX),200,200,1)
testX=testX.reshape(len(testX),200,200,1)
train_datagen=ImageDataGenerator(rescale=1./256)
history=model.fit_generator(train_datagen.flow(trainX,trainY),
                    steps_per_epoch=10,
                    epochs=10,
                    validation_data=train_datagen.flow(testX,testY),
                    validation_steps=1)
```

```python
from keras.preprocessing.image import ImageDataGenerator
dataaug= ImageDataGenerator(
                        rescale=1./256,
                        rotation_range=20,
                        width_shift_range=0.2,
                        height_shift_range=0.2,
                        shear_range=0.2,
                        zoom_range=0.2,
                        horizontal_flip=True,
                        fill_mode='nearest')

test_datagen=ImageDataGenerator(rescale=1./65536)
```

```python
history=model.fit_generator(dataaug.flow(trainX,trainY,shuffle=False),
                    steps_per_epoch=5,
                    epochs=50,
                    validation_data=test_datagen.flow(testX,testY,shuffle=False),
                    validation_steps=5)
```
