```python
from skimage import morphology
import skdemo
open_img = morphology.opening(img,morphology.square(5))
TO = open_img - img
skdemo.imshow_all(open_img,TO)
```

```python
from skimage import morphology
import skdemo
close_img = morphology.closing(img,morphology.square(5))
TO1 = img-close_img
skdemo.imshow_all(close_img,TO1)
```

```python
final = (TO+TO1)-img
```

```python
plt.imshow(final)
```

```python
from skimage import filters
th = filters.threshold_otsu(final)
plt.imshow(final>th)
```
