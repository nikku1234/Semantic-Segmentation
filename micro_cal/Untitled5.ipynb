```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters
from skimage import exposure

```

```python
img = cv2.imread('/Users/nikhil/Documents/GitHub/Semantic-Segmentation/micro_cal/inbreast/1/22427728_d713ef5849f98b6c_MG_R_ML_ANON.dcm.png',0)
plt.imshow(img,cmap='gray')

```

```python
#Apply tophat
kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat,cmap='gray')

```

```python
#Otsu's thresholding
val = filters.threshold_otsu(tophat)

hist, bins_center = exposure.histogram(tophat)

plt.figure(figsize=(9, 4))
plt.subplot(131)
plt.imshow(tophat, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.subplot(132)
plt.imshow(tophat < val, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.subplot(133)
plt.plot(bins_center, hist, lw=2)
plt.axvline(val, color='k', ls='--')

plt.tight_layout()
plt.show()
```
