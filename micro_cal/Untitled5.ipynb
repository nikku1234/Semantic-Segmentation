```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters
from skimage import exposure

img = cv2.imread('/content/mdb211.pgm',0)
plt.imshow(img,cmap='gray')

#Apply tophat
kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
plt.imshow(tophat,cmap='gray')

#Otsu's thresholding
val = filters.threshold_otsu(tophat)

hist, bins_center = exposure.histogram(tophat)

plt.figure(figsize=(9, 4))
plt.subplot(131)
plt.imshow(tophat, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.subplot(132)
plt.imshow(tophat < val, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.subplot(133)
plt.plot(bins_center, hist, lw=2)
plt.axvline(val, color='k', ls='--')

plt.tight_layout()
plt.show()
```
